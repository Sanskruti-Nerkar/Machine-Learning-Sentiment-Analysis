## Remove html tags

### Subtask:
Use regular expressions to remove any HTML tags that might be present in the text.


**Reasoning**:
Define a function to remove HTML tags from the text data using regular expressions as requested.



```python
import re

def remove_html_tags(text):
    """
    Removes HTML tags from a string.

    Args:
        text: The input string.

    Returns:
        The string with HTML tags removed.
    """
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

**Reasoning**:
Apply the `remove_html_tags` function to the 'text' column of each DataFrame as instructed by the subtask.



```python
df_train['text'] = df_train['text'].apply(remove_html_tags)
df_test['text'] = df_test['text'].apply(remove_html_tags)
df_validation['text'] = df_validation['text'].apply(remove_html_tags)

print("Train DataFrame after removing HTML tags:")
display(df_train.head())

print("\nTest DataFrame after removing HTML tags:")
display(df_test.head())

print("\nValidation DataFrame after removing HTML tags:")
display(df_validation.head())
```

## Remove punctuation

### Subtask:
Remove punctuation marks from the text.


**Reasoning**:
Define a function to remove punctuation and apply it to the 'text' column of the dataframes as instructed.



```python
import string

def remove_punctuation(text):
    """
    Removes punctuation from a string.

    Args:
        text: The input string.

    Returns:
        The string with punctuation removed.
    """
    return text.translate(str.maketrans('', '', string.punctuation))

df_train['text'] = df_train['text'].apply(remove_punctuation)
df_test['text'] = df_test['text'].apply(remove_punctuation)
df_validation['text'] = df_validation['text'].apply(remove_punctuation)

print("Train DataFrame after removing punctuation:")
display(df_train.head())

print("\nTest DataFrame after removing punctuation:")
display(df_test.head())

print("\nValidation DataFrame after removing punctuation:")
display(df_validation.head())
```

## Convert text to lowercase

### Subtask:
Convert all text to lowercase to ensure consistency.


**Reasoning**:
Define a function to convert text to lowercase and apply it to the text column of all three dataframes as instructed.



```python
def to_lowercase(text):
    """
    Converts a string to lowercase.

    Args:
        text: The input string.

    Returns:
        The lowercase string.
    """
    return text.lower()

df_train['text'] = df_train['text'].apply(to_lowercase)
df_test['text'] = df_test['text'].apply(to_lowercase)
df_validation['text'] = df_validation['text'].apply(to_lowercase)

print("Train DataFrame after converting to lowercase:")
display(df_train.head())

print("\nTest DataFrame after converting to lowercase:")
display(df_test.head())

print("\nValidation DataFrame after converting to lowercase:")
display(df_validation.head())
```

Remove common English stopwords that do not add significant meaning.


**Reasoning**:

## Remove emojis (optional)

### Subtask:
If emojis are present and not relevant to the analysis, remove them.


**Reasoning**:
Apply the remove_emojis function to the 'text' column of the train, test, and validation DataFrames and display the head of each DataFrame to show the results.



```python
def remove_emojis(text):
    """
    Removes emojis from a string using a regex pattern.

    Args:
        text: The input string.

    Returns:
        The string with emojis removed.
    """
    # Regex pattern to match a broad range of Unicode characters commonly used for emojis
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F700-\U0001F77F"  # Alchemical Symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r'', text)

# Apply the remove_emojis function to the 'text' column
df_train['text'] = df_train['text'].apply(remove_emojis)
df_test['text'] = df_test['text'].apply(remove_emojis)
df_validation['text'] = df_validation['text'].apply(remove_emojis)

print("Train DataFrame after removing emojis:")
display(df_train.head())

print("\nTest DataFrame after removing emojis:")
display(df_test.head())

print("\nValidation DataFrame after removing emojis:")
display(df_validation.head())
```

## Tokenize text

### Subtask:
Split the text into individual words or tokens.


**Reasoning**:
Import necessary NLTK modules, define a function to tokenize text, apply the function to the 'text' column of each dataframe, and display the head of each dataframe to show the new 'tokens' column as instructed.



```python
import nltk
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

from nltk.tokenize import word_tokenize

def tokenize_text(text):
    """
    Tokenizes a string into words.

    Args:
        text: The input string.

    Returns:
        A list of tokens.
    """
    return word_tokenize(text)

# Apply the tokenize_text function to the 'text' column
df_train['tokens'] = df_train['text'].apply(tokenize_text)
df_test['tokens'] = df_test['text'].apply(tokenize_text)
df_validation['tokens'] = df_validation['text'].apply(tokenize_text)

print("Train DataFrame after tokenization:")
display(df_train.head())

print("\nTest DataFrame after tokenization:")
display(df_test.head())

print("\nValidation DataFrame after tokenization:")
display(df_validation.head())
```

## Lemmatize tokens

### Subtask:
Convert tokens to their base or dictionary form (lemma).


**Reasoning**:
Define a function to lemmatize tokens using WordNetLemmatizer and apply it to the 'tokens' column of each dataframe, storing the results in a new 'lemmatized_tokens' column, and then display the head of each dataframe to show the results.



```python
from nltk.stem import WordNetLemmatizer

# Instantiate WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_tokens(tokens):
    """
    Lemmatizes a list of tokens.

    Args:
        tokens: A list of tokens.

    Returns:
        A list of lemmatized tokens.
    """
    return [lemmatizer.lemmatize(token) for token in tokens]

# Apply the lemmatize_tokens function to the 'tokens' column
df_train['lemmatized_tokens'] = df_train['tokens'].apply(lemmatize_tokens)
df_test['lemmatized_tokens'] = df_test['tokens'].apply(lemmatize_tokens)
df_validation['lemmatized_tokens'] = df_validation['tokens'].apply(lemmatize_tokens)

print("Train DataFrame after lemmatization:")
display(df_train.head())

print("\nTest DataFrame after lemmatization:")
display(df_test.head())

print("\nValidation DataFrame after lemmatization:")
display(df_validation.head())
```
